```python
import matplotlib.pyplot as plt
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
import numpy as np
# Generate a sample time series dataset with seasonality
date_rng = pd.date_range(start='2022-01-01', end='2023-12-31', freq='D')
seasonality = np.sin(2 * np.pi * date_rng.dayofyear / 365)
time_series_data = np.random.randn(len(date_rng)) + 0.2 * seasonality
# Create a time series DataFrame
time_series_df = pd.DataFrame(time_series_data, index=date_rng,
columns=['Value'])
# Decompose the time series into trend, seasonal, and residual components
result = seasonal_decompose(time_series_df['Value'], model='additive',
period=365)
# Plot the decomposed components
plt.figure(figsize=(12, 8))
plt.subplot(411)
plt.plot(time_series_df.index, result.observed, label='Observed')
plt.ylabel('Observed')
plt.title('Seasonal Decomposition of Time Series Data')
plt.subplot(412)
plt.plot(time_series_df.index, result.trend, label='Trend')
plt.ylabel('Trend')
plt.subplot(413)
plt.plot(time_series_df.index, result.seasonal, label='Seasonal')
plt.ylabel('Seasonal')
plt.subplot(414)
plt.plot(time_series_df.index, result.resid, label='Residual')
plt.ylabel('Residual')
plt.xlabel('Date')
plt.tight_layout()
# Show the plot
plt.grid(True)
plt.show()
```

```python
# Step 0: Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import zscore

# Step 1: Load the Titanic dataset
df = sns.load_dataset('titanic')


# Step 3: Replace 0/1 in 'Survived' column with labels and plot bar chart
df['Survived'] = df['Survived'].replace({0: 'Not Survived', 1: 'Survived'})

# Plot bar chart for Survived and Pclass
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")

plt.subplot(1, 2, 2)
sns.countplot(x='Pclass', data=df)
plt.title("Passenger Class Count")

plt.tight_layout()
plt.show()

# Step 4: Find Null Values and Fill with Mean
print("\n--- Null Value Count ---")
print(df.isnull().sum())

# Fill nulls in Age and Fare with their means
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Fare'].fillna(df['Fare'].mean(), inplace=True)

# Verify if nulls are filled
print("\n--- Nulls after Filling ---")
print(df[['Age', 'Fare']].isnull().sum())

# Step 5: Detect Outliers using Z-score
z_scores = zscore(df[['Age', 'Fare']])
abs_z_scores = np.abs(z_scores)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.boxplot(x=df['Age'])
plt.title("Boxplot for Age")

plt.subplot(1, 2, 2)
sns.boxplot(x=df['Fare'])
plt.title("Boxplot for Fare")

plt.tight_layout()
plt.show()
```

# **Python Descriptive Statistics â€“ Dispersion in Python**

# **1. Measures of Central Tendency:

**

```python
import numpy as np
from scipy import stats

# Sample data (replace this with your dataset)
data = np.array([10, 15, 20, 25, 30, 35, 40, 45, 50, 55])

# Central Tendency
mean = np.mean(data)
median = np.median(data)
mode = stats.mode(data).mode

# Dispersion
data_range = np.ptp(data)  # Peak-to-Peak Range
variance = np.var(data)  # Variance
std_deviation = np.std(data)  # Standard Deviation
iqr = np.percentile(data, 75) - np.percentile(data, 25)  # Interquartile Range (IQR)

# Print the results
print("Central Tendency:")
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")

print("\nDispersion:")
print(f"Range: {data_range}")
print(f"Variance: {variance}")
print(f"Standard Deviation: {std_deviation}")
print(f"IQR: {iqr}")
```

# **These are some basic linear algebra operations using Python and NumPy.

Tensors Gradients in python :
**

```python
import tensorflow as tf
x = tf.constant(2.0)
y = tf.constant(3.0)
with tf.GradientTape() as tape:
    tape.watch([x, y])
    z = x * y
grad_x, grad_y = tape.gradient(z, [x, y])
print(f"z = {z.numpy()}")
print(f"Gradient of z with respect to x: {grad_x.numpy()}")
print(f"Gradient of z with respect to y: {grad_y.numpy()}")
```

# **PyTorch Example:**

```python
import torch
x = torch.tensor(2.0, requires_grad=True)
y = torch.tensor(3.0, requires_grad=True)
z = x * y
z.backward()
grad_x = x.grad
grad_y = y.grad
print(f"z = {z.item()}")
print(f"Gradient of z with respect to x: {grad_x.item()}")
print(f"Gradient of z with respect to y: {grad_y.item()}")
```

# ***Outlier Detection***

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

data = pd.Series([160, 162, 158, 165, 161, 250])  # height in cm

# Boxplot
plt.boxplot(data)
plt.title("Univariate Outlier Detection - Boxplot")
plt.show()

# IQR method
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
outliers = data[(data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))]
print("Outliers:", outliers.values)
```
